#### 4. Кто 'уснувший' клиент?
Здесь следует проверить данные дат между покупками покупателей, у которых есть больше одной покупки

```python
count_purchase = ecom_df.groupby('customer_id', as_index=False).customer_id.value_counts()
count_purchase_filtred = count_purchase.query('count > 1')

ecom_filtred = ecom_df.query('customer_id in @count_purchase_filtred.customer_id')
ecom_filtred = ecom_filtred.sort_values(['customer_id', 'purchase_date'])
ecom_filtred['previus_purchase_date'] = ecom_filtred.groupby('customer_id')['purchase_date'].shift(1)
ecom_filtred['days_between_purchase'] = (
                                        ecom_filtred['purchase_date'] - ecom_filtred['previus_purchase_date']
                                        ).dt.days

ecom_filtred
```

```python
ecom_filtred['days_between_purchase'].describe()
```
